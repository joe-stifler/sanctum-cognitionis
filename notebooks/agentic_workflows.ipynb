```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
graph = nx.DiGraph()

# Add nodes (agents)
graph.add_node("Researcher")
graph.add_node("Analyst")
graph.add_node("Architect")

# Add edges (workflow)
graph.add_edge("Researcher", "Analyst", label="Preprocessed Text")
graph.add_edge("Analyst", "Architect", label="Formalized Algorithm")

# Draw the graph
pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_weight="bold")
edge_labels = nx.get_edge_attributes(graph, 'label')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)
plt.show()
```

## 1. Information Extraction Workflow

**Input:** The workflow starts with a research paper as input.

**Agent:** The Extractor Agent analyzes the paper and extracts relevant text, images, and tables.

**Outputs:** The agent produces extracted text, images, tables, and populates a task database with information for subsequent workflows.

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

# Input Content
graph.add_node("Input: Research Paper", shape="cylinder", style="filled", fillcolor="lightgreen")

# Agent
graph.add_node("Extractor Agent", shape="box", style="rounded", fillcolor="lightblue")

# Output Assets
graph.add_node("Output: Extracted Text", shape="cylinder", style="filled", fillcolor="lightyellow")
graph.add_node("Output: Extracted Images", shape="cylinder", style="filled", fillcolor="lightyellow")
graph.add_node("Output: Extracted Tables", shape="cylinder", style="filled", fillcolor="lightyellow")
graph.add_node("Output: Task Database", shape="cylinder", style="filled", fillcolor="lightcoral")

# Edges
graph.add_edge("Input: Research Paper", "Extractor Agent", label="Analyzes")
graph.add_edge("Extractor Agent", "Output: Extracted Text", label="Extracts")
graph.add_edge("Extractor Agent", "Output: Extracted Images", label="Extracts")
graph.add_edge("Extractor Agent", "Output: Extracted Tables", label="Extracts")
graph.add_edge("Extractor Agent", "Output: Task Database", label="Populates")

# Visualization
pos = nx.spring_layout(graph, seed=42)  # Seed for consistent layout
nx.draw(graph, pos, with_labels=True, node_size=3000, font_size=8, font_weight="bold", node_color=[n[1]['fillcolor'] for n in graph.nodes(data=True)])
edge_labels = nx.get_edge_attributes(graph, 'label')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)
plt.title("Information Extraction Workflow")
plt.show()
```

## 2. Task Assignment Workflow


**Input:** The Task Manager Agent receives the populated task database from the Information Extraction Workflow.

**Agent:** The Task Manager Agent queries the database, prioritizes tasks, and assigns them to appropriate Developer Agents.

**Output:** The output is a set of assigned tasks, ready for the Developer Agents to work on.

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

# Input
graph.add_node("Input: Task Database", shape="cylinder", style="filled", fillcolor="lightcoral")

# Agent
graph.add_node("Task Manager Agent", shape="box", style="rounded", fillcolor="lightblue")

# Output
graph.add_node("Output: Assigned Tasks", shape="cylinder", style="filled", fillcolor="lightyellow")

# Edges
graph.add_edge("Input: Task Database", "Task Manager Agent", label="Queries & Prioritizes")
graph.add_edge("Task Manager Agent", "Output: Assigned Tasks", label="Assigns to Developers")

# Visualization
pos = nx.spring_layout(graph, seed=42)
nx.draw(graph, pos, with_labels=True, node_size=3000, font_size=8, font_weight="bold", node_color=[n[1]['fillcolor'] for n in graph.nodes(data=True)])
edge_labels = nx.get_edge_attributes(graph, 'label')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)
plt.title("Task Assignment Workflow")
plt.show()
```

## 3. Code Development and Testing Workflow


```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

# Inputs
graph.add_node("Input: Assigned Tasks", shape="trapezium", style="filled", fillcolor="lightgreen")
graph.add_node("Input: Extracted Algorithm Description", shape="trapezium", style="filled", fillcolor="lightgreen")

# Agents
graph.add_node("Developer Agent", shape="box", style="rounded", fillcolor="lightblue")
graph.add_node("Testing Agent", shape="box", style="rounded", fillcolor="lightblue")

# Outputs
graph.add_node("Output: Devito DSL Code", shape="trapezium", style="filled", fillcolor="lightyellow")
graph.add_node("Output: Test Cases", shape="trapezium", style="filled", fillcolor="lightyellow")
graph.add_node("Output: Test Results", shape="trapezium", style="filled", fillcolor="lightyellow")

# Edges
graph.add_edge("Input: Assigned Tasks", "Developer Agent", label="Implements")
graph.add_edge("Input: Extracted Algorithm Description", "Developer Agent", label="Uses")
graph.add_edge("Developer Agent", "Output: Devito DSL Code", label="Generates")
graph.add_edge("Output: Devito DSL Code", "Testing Agent", label="Tests")
graph.add_edge("Testing Agent", "Output: Test Cases", label="Generates")
graph.add_edge("Testing Agent", "Output: Test Results", label="Produces")

# Visualization
pos = nx.spring_layout(graph, seed=42)

# Custom node shapes
node_shapes = set((aShape[1]["shape"] for aShape in graph.nodes(data=True)))
for aShape in node_shapes:
    nx.draw_networkx_nodes(graph, pos, node_shape=aShape, nodelist=[sNode[0] for sNode in filter(lambda x: x[1]["shape"]==aShape, graph.nodes(data=True))], node_size=3000, node_color=[n[1]['fillcolor'] for n in graph.nodes(data=True)])

nx.draw_networkx_edges(graph, pos)
edge_labels = nx.get_edge_attributes(graph, 'label')
nx.draw_networkx_labels(graph, pos, font_size=8, font_weight="bold")
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)
plt.title("Code Development and Testing Workflow")
plt.show()
```
